# Rocky III (1982)

A single-film case study. Rocky III opens with Rocky Balboa at the peak of his career, knocks him down through the defeat by Clubber Lang and the death of Mickey, and ends with the rematch victory. A classic triumph-fall-triumph shape. This notebook plots the emotional arc to see whether the pipeline reads that shape.

## 1. Setup

```python
import json
import matplotlib.pyplot as plt
from pathlib import Path

LOCAL_ROOT = Path('/Users/derinsavasan/Desktop/parsons-spring/thesis')
ARCS_DIR   = LOCAL_ROOT / 'docs/thesis-outputs/arcs'

film = {'slug': 'rocky-iii-1982', 'label': 'Rocky III (1982)'}
film['arc'] = json.loads((ARCS_DIR / f"{film['slug']}_arc.json").read_text())

print(f"{film['label']}  tokens={film['arc']['token_count']:,}")
```

## 2. The arc

Rocky opens the film as reigning champion: the first window sits at +0.90, with a peak of +1.53 around w6 as he rides through the title defenses and celebrity montage. The sharp drop at w2 (-1.61) reads as the early intrusion of Clubber Lang, who stalks the edges of the first act before the defeat. The arc pulls back up through the middle as Rocky trains, then crashes hard in the second half: w14 at -1.57 and w15 at -2.24, the deepest point in the film. That region maps onto the loss to Clubber and Mickey's death, which land back-to-back in the script. From w16 onward the arc climbs and stays positive through the end, which is the Apollo training montage and the rematch.

```python
plt.rcParams['font.family'] = 'Baskerville'

arc = film['arc']
xs = [w['position'] for w in arc['main_arc']]
ys = [w['z_score']  for w in arc['main_arc']]

fig, ax = plt.subplots(figsize=(6, 6))
ax.plot(xs, ys, color='black', linewidth=2, marker='o', markersize=4, zorder=3)
ax.fill_between(xs, ys, 0, where=[y >= 0 for y in ys], alpha=0.15, color='#999999', zorder=2)
ax.fill_between(xs, ys, 0, where=[y < 0 for y in ys],  alpha=0.25, color='#444444', zorder=2)
ax.axhline(0, color='gray', lw=1, alpha=0.5)
ax.set_xlim(0, 1)
ax.set_title(f"{film['label']} emotional arc", fontsize=13, fontweight='bold')
ax.set_xlabel('Story progress', fontsize=11)
ax.set_ylabel('Z-score', fontsize=11)
ax.tick_params(labelsize=10)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## 3. Arc in progress

Two snapshots of the arc as it is being drawn. The axes stay fixed so the remaining blank space shows how much of the film is still ahead.

```python
y_lim = (min(ys) - 0.4, max(ys) + 0.4)

def plot_partial(progress, title_suffix):
    cutoff = int(round(len(xs) * progress))
    px, py = xs[:cutoff], ys[:cutoff]

    fig, ax = plt.subplots(figsize=(10, 4))
    ax.plot(px, py, color='black', linewidth=2, marker='o', markersize=4, zorder=3)
    ax.fill_between(px, py, 0, where=[y >= 0 for y in py], alpha=0.15, color='#999999', zorder=2)
    ax.fill_between(px, py, 0, where=[y < 0 for y in py],  alpha=0.25, color='#444444', zorder=2)
    ax.axhline(0, color='gray', lw=1, alpha=0.5)
    ax.set_xlim(0, 1)
    ax.set_ylim(*y_lim)
    ax.set_title(f"{film['label']} emotional arc — {title_suffix}", fontsize=13, fontweight='bold')
    ax.set_xlabel('Story progress', fontsize=11)
    ax.set_ylabel('Z-score', fontsize=11)
    ax.tick_params(labelsize=10)
    ax.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

plot_partial(1/3, 'one-third in')
```

```python
plot_partial(0.9, 'near finishing')
```
